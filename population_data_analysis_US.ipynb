```python
# import the data visualization library plotnine
import plotnine as p9 
```

```python
pop_df = pd.read_csv('/Users/korakastengren/Python-Data-Analysis-and-Visualization-main/Kora_final_project/Population.csv')

# plot the year compared to the population change 
year_by_pop_change_plot = p9.ggplot(data=pop_df, mapping=p9.aes(x='Date',
y='Population Chg'))
year_by_pop_change_plot + p9.geom_point() # creates the plot

year_by_pop_change_plot + p9.geom_point() + p9.theme_bw() + p9.theme(axis_text_x = p9.element_text(angle=90))  #change x-axis labels to up/down 
```

```python
# make a line graph instead to help show trends
line_graph = p9.ggplot(pop_df, mapping = p9.aes(x="Date", y="Population Chg", group =1))
line_graph + p9.geom_line() + p9.theme_bw() + p9.theme(axis_text_x = p9.element_text(angle=90))  #change x-axis labels to up/down 
```

```python
# make a line graph of total population over time 
line_graph_pop = p9.ggplot(pop_df, mapping = p9.aes(x="Date", y="Population", group =1))
line_graph_pop + p9.geom_line() + p9.theme_gray() + p9.theme(axis_text_x = p9.element_text(angle=90))  #change x-axis labels to up/down and change theme to 'gray'
```

```python
# make a line graph of births over time 
line_graph_pop = p9.ggplot(pop_df, mapping = p9.aes(x="Date", y="Births", group =1))
line_graph_pop + p9.geom_line() + p9.theme_gray() + p9.theme(axis_text_x = p9.element_text(angle=90))  #change x-axis labels to up/down and change theme to 'gray'
```
